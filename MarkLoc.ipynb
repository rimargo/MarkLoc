```python
from tkinter import *
import random
from tkinter import messagebox

# Создание основного окна
newWindow = Tk()
newWindow.title('Настройка параметров')
c = Canvas(newWindow, width=150, height=30, bg="white",
           cursor="pencil")
ca = Canvas(newWindow, width=150, height=30, bg="white",
            cursor="pencil")
can = Canvas(newWindow, width=150, height=30, bg="white",
             cursor="pencil")
Label(newWindow, text="Наша программа 'Робот вышел погулять' позволяет").pack()
Label(newWindow, text="двигать робота, сканировать цвета при помощи датчиков и определять местоположение.").pack()
Label(newWindow, text="").pack()
Label(newWindow, text="В этом окне вы можете выбрать карту мира, считанную роботом, и запустить его.").pack()
Label(newWindow, text="Траектория движения состоит максимум из 5 движений робота").pack()
Label(newWindow, text="в направлениях: вверх, вниз, вправо, влево.").pack()
Label(newWindow, text="Управление роботом осуществляется с помощью соответствующих кнопок клавиатуры").pack()
Label(newWindow, text="после того, как запустите робота.").pack()
Label(newWindow, text="Важно: Если желаете обновить положение робота или карту мира, задать новую траекторию,").pack()
Label(newWindow, text="то закройте окно с полем и повторно нажмите кнопку 'Запуск робота'").pack()
Label(newWindow, text="Выберите карту мира:").pack()

#Объявление карт мира
def choise_1():
    global measurements
    measurements = ['green', 'blue', 'red', 'green', 'red']


def choise_2():
    global measurements
    measurements = ['blue', 'red', 'blue', 'red', 'blue']


def choise_3():
    global measurements
    measurements = ['red', 'green', 'red', 'blue', 'red']


var = IntVar()
var.set(0)
Radiobutton(newWindow, text="Интересная карта", command=choise_1, variable=var, value=0).pack()
c.create_rectangle(0, 0, 30, 30, fill="green")
c.create_rectangle(30, 0, 60, 30, fill="blue")
c.create_rectangle(60, 0, 90, 30, fill="pink")
c.create_rectangle(90, 0, 120, 30, fill="green")
c.create_rectangle(120, 0, 150, 30, fill="pink")
c.pack()

Radiobutton(newWindow, text="Тоже интересная карта", command=choise_2, variable=var, value=1).pack()
ca.create_rectangle(0, 0, 30, 30, fill="blue")
ca.create_rectangle(30, 0, 60, 30, fill="pink")
ca.create_rectangle(60, 0, 90, 30, fill="blue")
ca.create_rectangle(90, 0, 120, 30, fill="pink")
ca.create_rectangle(120, 0, 150, 30, fill="blue")
ca.pack()

Radiobutton(newWindow, text="Очень интересная карта", command=choise_3, variable=var, value=3).pack()
can.create_rectangle(0, 0, 30, 30, fill="pink")
can.create_rectangle(30, 0, 60, 30, fill="green")
can.create_rectangle(60, 0, 90, 30, fill="pink")
can.create_rectangle(90, 0, 120, 30, fill="blue")
can.create_rectangle(120, 0, 150, 30, fill="pink")
can.pack()


# Отрисовка
def draw():
    global canvas
    window = Toplevel(newWindow)
    window.title('Локализация робота')
    canvas = Canvas(window, width=500, height=500, bg="white",
                    cursor="pencil")

    canvas.create_rectangle(0, 0, 100, 100, fill="pink", outline="white")
    canvas.create_rectangle(0, 100, 100, 200, fill="green", outline="white")
    canvas.create_rectangle(0, 200, 100, 300, fill="pink", outline="white")
    canvas.create_rectangle(0, 300, 100, 400, fill="blue", outline="white")
    canvas.create_rectangle(0, 400, 100, 500, fill="green", outline="white")

    canvas.create_rectangle(100, 0, 200, 100, fill="green", outline="white")
    canvas.create_rectangle(100, 100, 200, 200, fill="pink", outline="white")
    canvas.create_rectangle(100, 200, 200, 300, fill="blue", outline="white")
    canvas.create_rectangle(100, 300, 200, 400, fill="green", outline="white")
    canvas.create_rectangle(100, 400, 200, 500, fill="blue", outline="white")

    canvas.create_rectangle(200, 0, 300, 100, fill="pink", outline="white")
    canvas.create_rectangle(200, 100, 300, 200, fill="blue", outline="white")
    canvas.create_rectangle(200, 200, 300, 300, fill="pink", outline="white")
    canvas.create_rectangle(200, 300, 300, 400, fill="blue", outline="white")
    canvas.create_rectangle(200, 400, 300, 500, fill="pink", outline="white")

    canvas.create_rectangle(300, 0, 400, 100, fill="blue", outline="white")
    canvas.create_rectangle(300, 100, 400, 200, fill="green", outline="white")
    canvas.create_rectangle(300, 200, 400, 300, fill="blue", outline="white")
    canvas.create_rectangle(300, 300, 400, 400, fill="pink", outline="white")
    canvas.create_rectangle(300, 400, 400, 500, fill="green", outline="white")

    canvas.create_rectangle(400, 0, 500, 100, fill="green", outline="white")
    canvas.create_rectangle(400, 100, 500, 200, fill="blue", outline="white")
    canvas.create_rectangle(400, 200, 500, 300, fill="pink", outline="white")
    canvas.create_rectangle(400, 300, 500, 400, fill="green", outline="white")
    canvas.create_rectangle(400, 400, 500, 500, fill="pink", outline="white")

    canvas.pack()
    window.bind("<Key>", snake)

#Параметры мира, движения и датчиков
world = [['red', 'green', 'red', 'blue', 'green'],
         ['green', 'red', 'blue', 'green', 'blue'],
         ['red', 'blue', 'red', 'blue', 'red'],
         ['blue', 'green', 'blue', 'red', 'green'],
         ['green', 'blue', 'red', 'green', 'red']]

pExact = 0.8
pOvershoot = 0.1
pUndershoot = 0.1
pHit = 0.6
pMiss = 0.2
measurements = ['green', 'blue', 'red', 'green', 'red']

# Вывод матрицы
def print_matrices(p):
    for row in p:
        print(' '.join(['{:.4f}'.format(elem) for elem in row]))

# Поиск максимума по вероятностям и где находится робот на его основе
def find(p):
    max_elem = p[0][0]
    for j in range(len(p)):
        for i in range(len(p[j] )):
            if p[j][i] > max_elem:
                 max_elem =  p[j][i]
    list_index_max =[(i, j) for i in range(len(p)) for j in range(len(p[i])) if p[i][j]  == max_elem]
    row, column = list_index_max[0]
    print('Робот находится в строке: ', row+1, ' и стоблце: ', column+1)


# Запуск окошка и рандомная точка
def select_startpoint(p):  
    i = random.randint(0, 4)
    j = random.randint(0, 4)
    p[i][j] = 1
    return i, j

# Запуск робота
def run():
    global p
    global k
    global glaz1
    global glaz2
    global rot
    global robot
    k = 0
    draw()
    p = [[0 for i in range(5)] for j in range(5)]
    data = select_startpoint(p)
    print("Старт:")
    print_matrices(p)
    y = data[0] * 100 + 30
    x = data[1] * 100 + 30
    robot = canvas.create_rectangle(x, y, x + 40, y + 40, fill="white", outline="black")
    glaz1 = canvas.create_oval(x + 20, y + 10, x + 30, y + 20, fill="black", outline="black")
    glaz2 = canvas.create_oval(x + 10, y + 10, x + 20, y + 20, fill="black", outline="black")
    rot = canvas.create_line(x + 7, y + 35, x + 33, y + 35)


b2 = Button(newWindow, bg="white", command=run, text="Запуск робота", activeforeground="blue",
            activebackground="pink", pady=10)
b2.pack()

# Сканирование мира
def sense(p, Z, k):
    p_new = [[0 for i in range(5)] for j in range(5)]
    s = 0
    for j in range(len(p)):
        for i in range(len(p[j])):
            hit = (Z[k] == world[j][i])
            p_new[j][i] = p[j][i] * (hit * pHit + (1 - hit) * pMiss)
            s += p_new[j][i]
    p_new = [[p_new[j][i] / s for i in range(5)] for j in range(5)]
    return p_new


# Вероятность для движений
def move_right(p):
    p_new = [[0 for i in range(5)] for j in range(5)]

    for j in range(len(p)):
        for i in range(len(p[j])):
            if i == 0:
                p_new[j][i] = p[j][i] * pUndershoot
                p_new[j][4] += p[j][(i - 1) % len(p[j])] * pExact + p[j][(i - 2) % len(p[j])] * pOvershoot
            elif i == 1:
                p_new[j][i] = p[j][i] * pUndershoot + p[j][i - 1] * pExact
                p_new[j][4] += p[j][(i - 2) % len(p[j])] * pOvershoot
            else:
                p_new[j][i] += p[j][i] * pUndershoot + p[j][i - 1] * pExact + p[j][i - 2] * pOvershoot


    return p_new


def move_left(p):
    p_new = [[0 for i in range(5)] for j in range(5)]

    for j in range(len(p)):
        for i in range(len(p[j])):
            if i == 4:
                p_new[j][i] = p[j][i] * pUndershoot
                p_new[j][0] += p[j][(i + 1) % len(p[j])] * pExact + p[j][(i + 2) % len(p[j])] * pOvershoot
            elif i == 3:
                p_new[j][i] = p[j][i] * pUndershoot + p[j][i + 1] * pExact
                p_new[j][0] += p[j][(i + 2) % len(p[j])] * pOvershoot
            else:
                p_new[j][i] = p[j][i] * pUndershoot + p[j][i + 1] * pExact + p[j][i + 2] * pOvershoot

    return p_new


def move_up(p):
    p_new = [[0 for i in range(5)] for j in range(5)]

    for j in range(len(p)):
        for i in range(len(p[j])):
            if j == 4:
                p_new[j][i] = p[j][i] * pUndershoot
                p_new[0][i] += p[(j + 1) % len(p)][i] * pExact + p[(j + 2) % len(p)][i] * pOvershoot
            elif j == 3:
                p_new[j][i] = p[j][i] * pUndershoot + p[j + 1][i] * pExact
                p_new[0][i] += p[(j + 2) % len(p)][i] * pOvershoot
            else:
                p_new[j][i] = p[j][i] * pUndershoot + p[j + 1][i] * pExact + p[j + 2][i] * pOvershoot

    return p_new


def move_down(p):
    p_new = [[0 for i in range(5)] for j in range(5)]

    for j in range(len(p)):
        for i in range(len(p[j])):
            if j == 0:
                p_new[j][i] = p[j][i] * pUndershoot
                p_new[4][i] += p[(j - 1) % len(p)][i] * pExact + p[(j - 2) % len(p)][i] * pOvershoot
            elif j == 1:
                p_new[j][i] = p[j][i] * pUndershoot + p[j - 1][i] * pExact
                p_new[4][i] += p[(j - 2) % len(p)][i] * pOvershoot
            else:
                p_new[j][i] = p[j][i] * pUndershoot + p[j - 1][i] * pExact + p[j - 2][i] * pOvershoot

    return p_new


# Отработка нажатия кнопок и расчет вероятностей
def snake(event):
    global p
    global k
    global robot
    global glaz1
    global glaz2
    global rot
    x = 0
    y = 0
    y_rob = canvas.coords(robot)[1]
    x_rob = canvas.coords(robot)[0]

    if k < 5:
        if event.keysym == 'Up':  
            if 30 < y_rob <= 430:
                y = -100
                tochka = canvas.create_text(x_rob + 15, y_rob + 15, anchor=W, font="DejavuSansLight", text=k + 1)
            p = sense(p, measurements, k)
            print('После сканирования:')
            print_matrices(p)
            p = move_up(p)
            print('После движения вверх:')
            print_matrices(p)
            find(p)
            k += 1

        if event.keysym == 'Left':
            if 30 < x_rob <= 430:
                x = -100
                tochka = canvas.create_text(x_rob + 15, y_rob + 15, anchor=W, font="DejavuSansLight", text=k + 1)
            p = sense(p, measurements, k)
            print('После сканирования:')
            print_matrices(p)
            p = move_left(p)
            print('После движения влево:')
            print_matrices(p)
            find(p)
            k += 1

        if event.keysym == 'Down':
            if 30 <= y_rob < 430:
                y = 100
                tochka = canvas.create_text(x_rob + 15, y_rob + 15, anchor=W, font="DejavuSansLight", text=k + 1)
            p = sense(p, measurements, k)
            print('После сканирования:')
            print_matrices(p)
            p = move_down(p)
            print('После движения вниз:')
            print_matrices(p)
            find(p)
            k += 1

        if event.keysym == 'Right':
            if 30 <= x_rob < 430:
                x = 100
                tochka = canvas.create_text(x_rob + 15, y_rob + 15, anchor=W, font="DejavuSansLight", text=k + 1)
            p = sense(p, measurements, k)
            print('После сканирования:')
            print_matrices(p)
            p = move_right(p)
            print('После движения вправо:')
            print_matrices(p)
            find(p)
            k += 1

        canvas.move(robot, x, y)
        canvas.move(glaz1, x, y)
        canvas.move(glaz2, x, y)
        canvas.move(rot, x, y)
    else:
        messagebox.showinfo("Внимание!",
                            "Отрисовка траектории движения робота окончена. Можете проверить полученные данные в поле вывода. Для перезапуска программы закройте окно с полем и нажмите кнопку 'Запуск робота'.")

newWindow.mainloop()
```
